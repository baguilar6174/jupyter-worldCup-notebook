## Modifying the dataframe

```python
# Create a column "earnings"

# Remove '$' from "Precio, Costo" columns
prices_df['Precio'] = prices_df['Precio'].str.replace("$","").astype("float")
prices_df['Costo'] = prices_df['Costo'].str.replace("$","").astype("float")

prices_df['Ganancias'] = prices_df['Precio'] - prices_df['Costo']

# Assigns values to the 'Margen' column based on a condition
prices_df['Margen'] = np.where(prices_df['Ganancias'] > 100, "Alto", "Bajo")

prices_df.head()
```

```python
# Drop columns

# Will remove the column named 'Hora de Cobro' from the DataFrame.
# The specification of axis=1 ensures that Pandas interprets 'Hora de Cobro' as a column name instead of a row index.

restaurant_df =  restaurant_df.drop(columns = ['Hora de Cobro'], axis = 1)
restaurant_df.info()
```

```python
# Rename columns

restaurant_df = restaurant_df.rename(columns = {"Atendió" : "Mesero"})
restaurant_df.info()
```

## Using Filters

```python
# Create a new dataframe

high_prices_df = prices_df[prices_df["Precio"] > 100]
high_prices_df.sort_values(by='Precio', ascending=True).head()
```

## Merging o Vlookup

Join two dataframes using left join

```python
restaurant_with_prices_df = restaurant_df.merge(prices_df, on = ["Producto"], how = "left")
restaurant_with_prices_df.head(20)
```
